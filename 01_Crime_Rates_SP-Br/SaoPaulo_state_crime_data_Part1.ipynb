```python
# Checking regions available

from selenium.webdriver.support.ui import Select
id_select = Select(driver.find_element_by_id('conteudo_ddlRegioes'))

reg_opt=[]
for opt in id_select.options:
    s=str(opt.get_attribute('innerHTML'))
    reg_opt.append(s)
print(reg_opt)
```

```python


#================TABLE 1 ============================

driver = webdriver.Chrome(options=chrome_options, executable_path=chrome_driver)
start = time.time()
df_crime_rate=[]
selec=Select(driver.find_element_by_id('conteudo_ddlAnos'))
selec.select_by_visible_text('Todos')

driver.page_source  #dynamic page update

#arr=range(0,len(regcity_opt))
arr=range(100,len(regcity_opt))

for j in range(len(arr)):
    i=arr[j]
    selec=Select(driver.find_element_by_id('conteudo_ddlRegioes'))
    selec.select_by_visible_text('%s' %regcity_opt[i][0])
    
    selec=Select(driver.find_element_by_id('conteudo_ddlMunicipios'))
    selec.select_by_visible_text('%s' %regcity_opt[i][1])
    
    python_button = driver.find_element_by_id('conteudo_btnTaxa') #FHSU
    python_button.click()
    
    driver.page_source

    if check_exists_by_id('conteudo_gridTaxa')==True:
        table = driver.find_element_by_id('conteudo_gridTaxa')
        tablehtml = table.get_attribute('outerHTML')
        df = pd.read_html(tablehtml,thousands = None)
        if j==0:
            df_crime_rate=pd.DataFrame(df[0])
            df_crime_rate['Cidade'] = pd.Series(np.repeat('%s'% regcity_opt[i][1],df1.shape[0] ))
            df_crime_rate['Regiao'] = pd.Series(np.repeat('%s'% regcity_opt[i][0],df1.shape[0] ))
        elif j!=0:
            df1=pd.DataFrame(df[0])
            df1['Cidade'] = pd.Series(np.repeat('%s'% regcity_opt[i][1],df1.shape[0] ))
            df1['Regiao'] = pd.Series(np.repeat('%s'% regcity_opt[i][0],df1.shape[0] ))
            df_crime_rate=df_crime_rate.append(df1) 
    
    elif check_exists_by_id('conteudo_gridTaxa')==False:
        if j==0: #if frist row do not have data available
            df_crime_rate=pd.DataFrame(columns=['Ano', 'Homicídio Doloso por 100 mil habitantes',
           'Furto por 100 mil habitantes', 'Roubo por 100 mil habitantes','Furto e Roubo de Veículo por 100 mil habitantes','Furto por 100 mil veículos', 'Roubo por 100 mil veículos',
           'Furto e Roubo de Veículo por 100 mil veículos']  )
            df_crime_rate.loc[0] = ["None","None","None","None","None","None","None","None"]
            df_crime_rate ['Cidade'] = pd.Series(np.repeat('%s'% regcity_opt[i][1],1))
            df_crime_rate ['Regiao'] = pd.Series(np.repeat('%s'% regcity_opt[i][0],1))
        else:
            df2=pd.DataFrame(columns=['Ano', 'Homicídio Doloso por 100 mil habitantes',
           'Furto por 100 mil habitantes', 'Roubo por 100 mil habitantes','Furto e Roubo de Veículo por 100 mil habitantes','Furto por 100 mil veículos', 'Roubo por 100 mil veículos',
           'Furto e Roubo de Veículo por 100 mil veículos'])
            df2.loc[0] = ["None","None","None","None","None","None","None","None"]
            df2['Cidade'] = pd.Series(np.repeat('%s'% regcity_opt[i][1],1))
            df2['Regiao'] = pd.Series(np.repeat('%s'% regcity_opt[i][0],1))
            df_crime_rate=df_crime_rate.append(df2) 


elapsed_time_fl = (time.time() - start) 
print("elapsed_time_fl:     %f    s" % elapsed_time_fl )


export_csv = df_crime_rate.to_csv (r'C:\Users\Dalciana\Python_DBW_files\01_proj_crimes_sp\df_crime_rate.csv', index = None, header=True) #Don't forget to add '.csv' at the end of the path
df_crime_rate
```
